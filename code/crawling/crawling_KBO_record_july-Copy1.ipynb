## Crawling - KBO 기록 7월 추가

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
```

```python
# 크롬드라이버 경로 설정
path = 'C:/Users/dhxog/Downloads/chromedriver_win32 (4)/chromedriver'
```

### 경기 일정 크롤링

```python
# 경기일정 url
url = 'https://www.koreabaseball.com/Schedule/Schedule.aspx'

# 크롬드라이버 시작
driver = webdriver.Chrome(path)
driver.get(url)
time.sleep(10)

def setMonth(month):
    month = driver.find_element_by_xpath("//option[@value='" + int(month) + "']")
    month.click()

setMonth(7)
```

```python
# 크롤링 시작
table_df_list = []

for i in range(1, 9):
    
    # 해당 페이지 테이블 크롤링
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.select('#cphContents_cphContents_cphContents_udpContent > div.record_result > table')
    table_html = str(table)
    table_df = pd.read_html(table_html)
    table_df = table_df[0]
    table_df_list.append(table_df)
    
    time.sleep(10)
    
    # 다음 페이지로 이동
    if i%5==1:
        pagexpath = '//*[@id="cphContents_cphContents_cphContents_ucPager_btnNo2"]'
    elif i%5==2:
        pagexpath = '//*[@id="cphContents_cphContents_cphContents_ucPager_btnNo3"]'
    elif i%5==3:
        pagexpath = '//*[@id="cphContents_cphContents_cphContents_ucPager_btnNo4"]'
    elif i%5==4:
        pagexpath = '//*[@id="cphContents_cphContents_cphContents_ucPager_btnNo5"]'
    else:
        pagexpath = '//*[@id="cphContents_cphContents_cphContents_ucPager_btnNext"]'
    try:
        driver.find_element_by_xpath(pagexpath).click()
    except:
        break
    
    time.sleep(10)

# 크롬드라이버 종료
driver.quit()
```

```python
# 타자 7월 테이블
hitter_july = pd.concat(table_df_list)
hitter_july.reset_index(drop=True, inplace=True)
hitter_july
```

```python
hitter_july.to_csv('../../data/kbo_record_hitter_july.csv')
```
