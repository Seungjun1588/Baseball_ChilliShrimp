```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
import lxml
```

```python
# 7월 데이터 크롤링
url = 'https://sports.news.naver.com/kbaseball/schedule/index.nhn?month=07&year=2020'
req = requests.get(url)
html = req.text
```

```python
table_july = pd.read_html(html)
table_july
```

```python
# 8월 데이터 크롤링
url = 'https://sports.news.naver.com/kbaseball/schedule/index.nhn?month=08&year=2020'
req = requests.get(url)
html = req.text
```

```python
table_august = pd.read_html(html)
table_august
```

```python
# 9월 데이터 크롤링
url = 'https://sports.news.naver.com/kbaseball/schedule/index.nhn?month=09&year=2020'
req = requests.get(url)
html = req.text
```

```python
table_sept = pd.read_html(html)
table_sept
```

```python
# 10월 데이터 크롤링
url = 'https://sports.news.naver.com/kbaseball/schedule/index.nhn?month=10&year=2020'
req = requests.get(url)
html = req.text
```

```python
table_oct = pd.read_html(html)
table_oct
```

```python
# 데이터프레임 생성
df = pd.DataFrame()
for i in range(1, len(table_july)):
    df = df.append(table_july[i])
for i in range(1, len(table_august)):
    df = df.append(table_august[i])
for i in range(1, len(table_sept)):
    df = df.append(table_sept[i])
for i in range(1, len(table_oct)):
    df = df.append(table_oct[i])
```

```python
df
```

```python
# 컬럼 이름 생성
df.columns = ['날짜', '시간', '경기', '중계/기록', '중계방송사', '구장', '알림받기']
df
```

```python
cdf = df.copy()
```

```python
# 필요없는 행 제거
cdf = cdf[cdf.경기 != '프로야구 경기가 없습니다.']
cdf = cdf[cdf.중계방송사 != '해당 경기는 현지 사정으로 취소되었습니다.']
```

```python
# 필요 없는 열 제거
cdf.drop(['시간', '중계/기록', '중계방송사', '알림받기'], axis=1, inplace=True)
```

```python
# NA 제거
cdf = cdf.dropna()
```

```python
# 인덱스 리셋
cdf.reset_index(drop=True, inplace=True)
cdf
```

```python
# 연 컬럼 생성
year = [2020 for _ in range(len(cdf['날짜']))]
cdf['year'] = year
```

```python
# 월, 일 컬럼 생성
month = []
day = []
for i in range(len(cdf['날짜'])):
    string = cdf.loc[i, '날짜']
    pattern = re.compile('\(.\)')
    ls = re.sub(pattern, '', string).split('.')
    month.append(int(ls[0].strip()))
    day.append(int(ls[1].strip()))

cdf['month'] = month
cdf['day'] = day
```

```python
# 텍스트 전처리 함수 선언
def cleanText(string):
    pattern = re.compile('경기취소')
    string = re.sub(pattern, '', string)
    pattern = re.compile('\d+:\d+')
    string = re.sub(pattern, 'VS', string)
    return string
```

```python
# 왼쪽 팀, 오른쪽 팀 컬럼 생성
team_lft = []
team_rht = []
for i in range(len(cdf['경기'])):
    a = cdf.loc[i, '경기']
    a = cleanText(a)
    teams = a.split('VS')
    team_lft.append(teams[0].strip())
    team_rht.append(teams[1].strip())

cdf['team_visit'] = team_lft
cdf['team_home'] = team_rht
```

```python
# 경기 결과 컬럼 생성
score = []
for i in range(len(cdf)):
    string = cdf.loc[i, '경기']
    if 'VS' in string:
        v = ''
        score.append(v)
    else:
        pattern = re.compile('\d+:\d+')
        v = re.search(pattern, string)
        v = v.group()
        score.append(v)

cdf['score'] = score
```

```python
# 날짜, 경기 컬럼 삭제
cdf.drop(['날짜', '경기'], axis=1, inplace=True)
```

```python
# 컬럼 순서 변경
cdf = cdf.reindex(columns=['year', 'month', 'day', 'team_visit', 'team_home', '구장', 'score'])
```

```python
# 크롤링 시점 이전 데이터만 저장
cdf = cdf.loc[cdf['score']!='']
cdf
```

```python
# 저장
cdf.to_csv('../../data/kbo_schedule_before_0915.csv')
```
