```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
import lxml
```

```python
# 5월 데이터 크롤링
url = 'https://sports.news.naver.com/kbaseball/schedule/index.nhn?month=05&year=2020'
req = requests.get(url)
html = req.text
```

```python
table_may = pd.read_html(html)
table_may
```

```python
# 6월 데이터 크롤링
url = 'https://sports.news.naver.com/kbaseball/schedule/index.nhn?month=06&year=2020'
req = requests.get(url)
html = req.text
```

```python
table_june = pd.read_html(html)
table_june
```

```python
# 7월 데이터 크롤링
url = 'https://sports.news.naver.com/kbaseball/schedule/index.nhn?month=07&year=2020'
req = requests.get(url)
html = req.text
```

```python
table_july = pd.read_html(html)
table_july
```

```python
# 데이터프레임 생성
df = pd.DataFrame()
for i in range(1, len(table_may)):
    df = df.append(table_may[i])
for i in range(1, len(table_june)):
    df = df.append(table_june[i])
for i in range(1, len(table_july)):
    df = df.append(table_july[i])
```

```python
df
```

```python
df.columns = ['날짜', '시간', '경기', '중계/기록', '중계방송사', '구장', '알림받기']
df
```

```python
cdf = df.copy()
cdf = cdf[cdf.경기 != '프로야구 경기가 없습니다.']
cdf = cdf[cdf.중계방송사 != '해당 경기는 현지 사정으로 취소되었습니다.']
cdf.drop(['시간', '중계/기록', '중계방송사', '알림받기'], axis=1, inplace=True)
cdf = cdf.dropna()
```

```python
cdf.reset_index(drop=True, inplace=True) # 인덱스 리셋
cdf
```

```python
# 연 컬럼 생성
year = [2020 for _ in range(len(cdf['날짜']))]
cdf['year'] = year
```

```python
# 월 컬럼 생성
month = []
day = []
for i in range(len(cdf['날짜'])):
    string = cdf.loc[i, '날짜']
    pattern = re.compile('\(.\)')
    ls = re.sub(pattern, '', string).split('.')
    month.append(int(ls[0].strip()))
    day.append(int(ls[1].strip()))

cdf['month'] = month
cdf['day'] = day
```

```python
# 시범경기 삭제
cdf = cdf[5:]
```

```python
cdf.reset_index(drop = True, inplace = True)
```

```python
# 텍스트 전처리 함수 선언
def cleanText(string):
    pattern = re.compile('경기취소')
    string = re.sub(pattern, '', string)
    pattern = re.compile('\d+:\d+')
    string = re.sub(pattern, 'VS', string)
    return string
```

```python
# 왼쪽 팀, 오른쪽 팀 컬럼 생성
team_lft = []
team_rht = []
for i in range(len(cdf['경기'])):
    a = cdf.loc[i, '경기']
    a = cleanText(a)
    teams = a.split('VS')
    team_lft.append(teams[0].strip())
    team_rht.append(teams[1].strip())

# 데이터프레임에 컬럼 추가
cdf['team_visit'] = team_lft
cdf['team_home'] = team_rht

# 경기 컬럼 삭제
cdf.drop('경기', axis=1, inplace=True)
```

```python
# 컬럼 순서 변경
cdf = cdf.reindex(columns=['year', 'month', 'day', 'team_visit', 'team_home', '구장'])
cdf
```
