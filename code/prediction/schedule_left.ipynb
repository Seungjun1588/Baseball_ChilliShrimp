```python
import pandas as pd
```

```python
schedule = pd.read_csv("C:\\Users\\dhxog\\Desktop\\ESC_summer\\Baseball_ChilliShrimp\\data/schedule.csv")
before = pd.read_csv("C:\\Users\\dhxog\\Desktop\\ESC_summer\\Baseball_ChilliShrimp\\data/before_0719.csv")
```

```python
schedule = schedule.rename(columns = {'APP_T_ID':'VS_T_ID'})
before = before.rename(columns = {'APP_T_ID':'VS_T_ID'})
```

```python
schedule = pd.get_dummies(schedule, columns=['VS_T_ID', 'H_A'])
before = pd.get_dummies(before, columns=['VS_T_ID', 'H_A'])
```

```python
schedule = schedule.groupby(['year', 'month', 'T_ID']).sum()
```

```python
schedule
```

```python
schedule = schedule.drop(columns = ['H_A_AWAY', 'H_A_HOME'])
```

```python
schedule
```

```python
before = before.groupby(['year', 'month', 'T_ID']).sum()
```

```python
before = before.drop(columns=['H_A_AWAY', 'H_A_HOME'])
```

```python
gamed = before.groupby(['T_ID']).sum()
```

```python
future = schedule.groupby(['T_ID']).sum()
```

```python
gamed
```

```python
future
```

```python
not_suspended_game = gamed + future
```

```python
all_game = not_suspended_game.copy() * 0
```

```python
all_game = all_game + 16
```

```python
game_left = all_game - not_suspended_game
```

```python
game_left[game_left == 16] = 0
```

```python
game_left
```

```python
schedule_left = schedule[:-10].append(schedule[-10:] + game_left)
```

```python
schedule_left.to_csv('../../data/schedule_left.csv')
```
