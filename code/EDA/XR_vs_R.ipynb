#### XR vs R

```python
import numpy as np
import pandas as pd
```

```python
batter = pd.read_csv("C:/Users/njj06/Desktop/ESC_summer/Baseball_ChilliShrimp/data/batter_tidy_XR.csv")
pitcher = pd.read_csv("C:/Users/njj06/Desktop/ESC_summer/Baseball_ChilliShrimp/data/pitcher.csv")
```

```python
df1 = batter.loc[:,['year', 'month','T_ID', 'XR']]
df2 = pitcher.loc[:,['year', 'month','T_ID','R']]
```

```python
team = df1['T_ID'].unique()
year = df1['year'].unique()
month = df1['month'].unique()
```

```python
Y=[]
M=[]
T=[]
xr=[]
for y in year:
    for m in month:
        for t in team:
            xr.append(df1.loc[(df1['year'] == y) & (df1['month'] == m) & (df1['T_ID'] == t), 'XR'].sum())
            Y.append(y)
            M.append(m)
            T.append(t)
            
df = pd.DataFrame()
df['year'] = Y
df['month'] = M
df['T_ID'] = T
df['XR'] = xr
```

```python
r=[]
for y in year:
    for m in month:
        for t in team:
            r.append(df2.loc[(df2['year'] == y) & (df2['month'] == m) & (df2['T_ID'] == t), 'R'].sum())
df['R'] = r
```

```python
df = df.loc[:359]
```
