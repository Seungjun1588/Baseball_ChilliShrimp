## 파생변수 하나의 데이터셋으로 합치기

```python
import pandas as pd
```

### batter

```python
data = pd.read_csv('../../data/batter_tidy.csv')
data
```

```python
df1 = pd.read_csv('../../data/batter_tidy_볼넷당삼진_도루성공율.csv', index_col = 0)
df1
```

```python
data['KBB'] = df1['KBB']
data['SB_rate'] = df1['SB_rate']
data
```

```python
df2 = pd.read_csv('../../data/batter_tidy_AVG_OBP_SLG_OPS.csv', index_col = 0)
df2
```

```python
data['AVG'] = df2['AVG']
data['OPB'] = df2['OBP']
data['SLG'] = df2['SLG']
data['OPS'] = df2['OPS']
data
```

```python
df3 = pd.read_csv('../../data/batter_tidy_XR.csv', index_col = 0)
df3
```

```python
data['XR'] = df3['XR']
data
```

```python
data.to_csv('../../data/batter_all.csv')
```
