```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('C:/Users/seungjun/Desktop/baseball/data/raw_pitcher.csv')
```

```python
data1 = data.copy()
```

```python
data1
```

```python
data2 = data1[data['year']==2020]
```

```python
data2 = data2.drop(['BABIP'], axis=1)
```

```python
data2
```

```python
df = data2.groupby('P_ID').sum()
```

```python
df
```

```python
df['SF'] = df['PA'] - df['AB'] - df['BB'] - df['HP']
```

```python
df['BABIP'] = ( df['HIT'] - df['HR'] ) / (df['AB'] - df['KK'] - df['HR'] - df['SF'])
```

```python
df
```

```python
sum(np.isnan(df['BABIP']))
```

```python
sum(np.isinf(df['BABIP']))
```

```python
df['BABIP'] = df['BABIP'].fillna(0)
sum(np.isnan(df['BABIP']))
```

```python
babip_2020 = df[['BABIP']]
babip_2020['P_ID'] = babip_2020.index
```

```python
babip_2020 = babip_2020.reset_index(drop = True)
```

```python
babip_2020.to_csv('C:/Users/seungjun/Desktop/baseball/data/pitcher_babip_2020.csv')
```
