## Preprocessing

### 파생변수 생성하기 - DER

**공식** : DER = (타석 - 안타 - 삼진 - 볼넷 - 몸에맞는볼 - 에러출루) + (타석 - 홈런 - 삼진 - 볼넷 - 몸에맞는공)

**해석** : '인플레이 타구를 수비수들이 아웃으로 잘 잡아냈는가?'를 나타내는 지표

```python
import pandas as pd
```

```python
data = pd.read_csv('../../data/pitcher_all.csv', index_col=0)
data
```

```python
df = data.copy()
```

```python
df['DER'] = (df['PA'] - df['HIT'] - df['KK'] - df['BB'] - df['HP'] - df['ERR']) + (df['PA'] - df['HR'] - df['KK'] - df['BB'] - df['HP'])
```

```python
# 기존 pitcher_all.csv 파일 파생변수 추가하여 업데이트
df.to_csv('../../data/pitcher_all.csv')
```
