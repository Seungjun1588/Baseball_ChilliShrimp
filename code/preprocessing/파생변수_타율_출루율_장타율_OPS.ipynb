### 타율 (AVG)
 
타율 = 안타수 / 타수
- 안타 = HIT
- 타수 = AB

```python
# 타율
df['AVG'] = df['HIT'] / df['AB']
```

```python
df[['AB', 'HIT', 'AVG']]
```

```python
sum(np.isnan(df['AVG']))
```

```python
sum(df['AVG'] == 0)
```

분모에 들어가는 타수(AB)가 0인 경우 AVG 값이 NaN으로 들어가는데,  
처음엔 이 NaN을 0으로 채우려고 함  
$\rightarrow$ 하지만, 분모인 타수(AB)는 0이 아닌데 분자인 안타(HIT)가 0이어서 AVG 값이 0으로 들어간 경우와 동일하게 취급된다는 점에서 문제  
$\rightarrow$ NA imputation 어떻게 해야할지 고민 : 안타(HIT)이 0일 때 분모에 엄청엄청 작은 수를 더해서 나눗셈이 되도록 하기..? 의견 필요..!

### 출루율 (OBP)

출루율 = (안타+볼넷+사구) / (타수+볼넷+사구+희생플라이)
- 안타 = HIT
- 볼넷 = BB
- 사구 = HP
- 타수 = AB
- 희생플라이 = SF

```python
# 출루율
df['OBP'] = (df['HIT'] + df['BB'] + df['HP']) / (df['AB'] + df['BB'] + df['HP'] + df['SF'])
```

```python
df[['HIT', 'BB', 'HP', 'AB', 'SF', 'OBP']]
```

```python
sum(df['OBP'] == 0)
```

```python
sum(np.isnan(df['OBP']))
```

### 장타율 (SLG)

SLG = (단타개수x1 + 2루타개수x2 + 3루타개수x3 + 홈런개수x4) / 타수
- 단타 = HIT-H2-H3-HR
- 2루타 = H2
- 3루타 = H3
- 홈런 = HR
- 타수 = AB

```python
# 장타율
df['SLG'] = ((df['HIT'] - df['H2'] - df['H3'] - df['HR'])*1 + df['H2']*2 + df['H3']*3 + df['HR']*4) / df['AB']
```

```python
sum(np.isnan(df['SLG']))
```

```python
sum(df['SLG']==0)
```

출루율과 장타율도 타율과 마찬가지로 NA imputation 어떻게 할지 고민  
특히 이들은 OPS 구할 때 사용되기 때문에 타율은 NaN으로 남겨놓더라도 출루율이랑 장타율은 NA imputation 꼭 해야 할 것 같다는 개인적인 생각

### OPS

OPS = 출루율 + 장타율

```python
df['OPS'] = df['OBP'] + df['SLG']
```
