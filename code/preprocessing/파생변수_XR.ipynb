XR(eXtrapolated Runs)  
=(0.5 * 1루타) + (0.72 * 2루타) + (1.04 * 3루타) + (1.44 * 홈런) + 0.34 * (사구-고의사구-4구) + (0.25 * 고의사구) + (0.18 * 도루) - (0.32 * 도루실패) - 0.09 * (타수 - 안타 - 삼진) - (0.098 * 삼진) - (0.37 * 병살타) + (0.37 * 희생플라이) + (0.04 * 희생번트)


출처: https://peria1024.tistory.com/20 [김에밀의 프로야구 이야기]

```python
df = data.copy()
```

```python
# 1루타
singles = df['HIT'] - df['HR'] - df['H2'] -df['H3'] 
```

```python
df['XR'] = (0.5*singles
            + 0.72*df['H2'] 
            + 1.04*df['H3'] 
            + 1.44*df['HR'] 
            + 0.34*(df['HP'] - df['IB'] + df['BB']) 
            + 0.25*df['IB'] 
            + 0.18*df['SB'] 
            - 0.32*df['CS'] 
            - 0.09*(df['AB'] - df['HIT'] - df['KK']) 
            - 0.098*df['KK'] 
            - 0.37*df['GD'] 
            + 0.27*df['SF'] 
            + 0.04*df['SH'])
```
