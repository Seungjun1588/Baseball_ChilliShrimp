```python
df = data.copy()
```

```python
df2 = df.drop(['BABIP'], axis=1)
```

```python
df2
```

```python
df2 = df2.groupby('P_ID').sum()
```

```python
df2['SF'] = df2['PA'] - df2['AB'] - df2['BB'] - df2['HP']
```

```python
df2['BABIP'] = ( df2['HIT'] - df2['HR'] ) / (df2['AB'] - df2['KK'] - df2['HR'] - df2['SF'])
```

```python
df2
```

```python
sum(np.isnan(df2['BABIP']))
```

```python
sum(np.isinf(df2['BABIP']))
```

```python
babip = df2[['BABIP']]
babip['P_ID'] = babip.index
```

```python
babip = babip.reset_index(drop = True)
```
