# Preprocessing

## 파생변수 생성하기 - 타율, 출루율, 장타율, OPS (with fitcher.csv)

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../../data/pitcher.csv')
data
```

```python
df = data.copy()
```

```python
# 타율
df['AVG'] = df['HIT'] / df['AB']
```

```python
df['AVG'] = df['AVG'].fillna(0)
```

```python
# 출루율
df['OBP'] = (df['HIT'] + df['BB'] + df['HP']) / (df['AB'] + df['BB'] + df['HP'] + df['SF'])
```

```python
df['OBP'] = df['OBP'].fillna(0)
```

```python
# 장타율
df['SLG'] = ((df['HIT'] - df['H2'] - df['H3'] - df['HR'])*1 + df['H2']*2 + df['H3']*3 + df['HR']*4) / df['AB']
```

```python
df['SLG'] = df['SLG'].fillna(0)
```

```python
# OPS
df['OPS'] = df['OBP'] + df['SLG']
```

```python
df.to_csv('../../data/pitcher_AVG_OBP_SLG_OPS.csv')
```
